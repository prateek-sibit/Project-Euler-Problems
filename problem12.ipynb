The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 **1: 1**
 
 **3: 1,3**
 
 
 **6: 1,2,3,6**

**10: 1,2,5,10**

**15: 1,3,5,15**

**21: 1,3,7,21**

**28: 1,2,4,7,14,28**

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import numpy as np

def len_divisors(number):
    nums = np.arange(start=1, stop=number-1, step=1)
    return len(nums[number % nums == 0])
```

```python
def nth_triangle(n):
    nums = np.arange(1, n)
    return np.sum(nums)
```

```python
'''
Algorithm : 
1. Generate triangle number
2. triangle_nth = 1 + 2 + 3 + 4 + ..... n
3. if len_divisors(triangle_nth) > 100 -> print, break loop
4. else iterate triangle_nth
'''
i = 1
triangle_num = nth_triangle(i)
while True:
    
    if len_divisors(triangle_num) > 499:
        print(triangle_num)
        break
    i += 1
    triangle_num = nth_triangle(i)
```
