# Problem 17
Number letter counts

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


**NOTE**: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
def number_name_lt20(number):
    return ones[number]

def number_name_20100(number):
    # Case 1 -> Simple 10s numbers
    if int(str(number)[1]) == 0:
        return tens[int(str(number)[0])]
    else:
        return tens[int(str(number)[0])] + ones[int(str(number)[1])]  
```

```python
ones = {1:'one', 2:'two', 3:'three', 4:'four', 5:'five',
         6:'six', 7:'seven', 8:'eight', 9:'nine', 10:'ten',
         11:'eleven', 12:'twelve', 13:'thirteen', 14:'fourteen', 15:'fifteen',
         16:'sixteen', 17:'seventeen', 18:'eighteen', 19:'nineteen', 20:'twenty'}
tens = {2:'twenty', 3:'thirty', 4:'forty', 5:'fifty', 6:'sixty', 7:'seventy', 8:'eighty', 9:'ninety'}
'''
Rule for writing out a number
1. For numbers < 20 -> Take ones[number]
2. For 20 < numbers < 100 -> Take first digit and take tens[digit one] + ones[digit two]
3. For numbers > 100 -> ones[digit one] + 'hundred and' + 
    2 Cases arise
    if number[1:] < 20 -> ones[digit one] + 'hundred and' + ones[number[1:]]
    elif number[1:] >20 -> ones[digit one] + 'hundred and' +  tens[digit two] + ones[digit three]
'''
def number_name(number):
    '''
    Input : Integer number
    Output : What the number will be pronounced as as per British Usage
    '''
    if number <= 20:
        return number_name_lt20(number)
    
    elif number > 20 and number < 100:
        return number_name_20100(number)
        
    elif number >= 100 and number < 1000:
        # Three cases arise
        
        # Base case -> Simple 100s number
        if str(number)[1:] == '00':
            return ones[int(str(number)[0])] + 'hundred'
        else:
            # Case 1 
            if int(str(number)[1:]) <= 20:
                return ones[int(str(number)[0])] + 'hundredand' + number_name_lt20(int(str(number)[1:]))
            # Case 2
            elif int(str(number)[1:]) > 20:
                return ones[int(str(number)[0])] + 'hundredand' + number_name_20100(int(str(number)[1:]))
            
    elif number == 1000:
        return 'onethousand'
```

```python
letters_used = 0
for i in range(1, 1001):
    print(number_name(i))
    letters_used += len(number_name(i))
print('Total Letters Used : ', letters_used)
```
