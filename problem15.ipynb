# Problem 15

**Lattice paths** : Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.
How many such routes are there through a 20×20 grid?

```python
# For an (n, n) Grid the corresponding search matrix is of size (n+1, n+1)
import numpy as np # To create the zeros matrix
```

```python
n = 2
grid = np.ndarray.tolist(np.zeros((n+1, n+1))) 
# Set Grid start and end point as 1 for identifier
grid[0][0] = -1
grid[n][n] = 1
print(grid)
```

```python
# At each matrix[i, j] two options exist -> Right : [i, j+1] or Down : [i+1, j]
# Traverse all paths possible and keep appending to the unique paths list
# At each matrix[i, j] make a random choice 
'''
Random choice algorithm : 
1. If randint == 0 : Choose Right
2. If randint == 1 : Choose Down
3. if right not possible choose down and vice versa
4. Do choosing and updation till matrix[i, j] != 1
'''
paths = []

i, j = 0, 0 # (i, j) represent the current matrix element coordinates
current_path = [(i, j)]
    
'''
THE BELOW CODE FINDS A ROUTE FROM THE START TO THE END
current_path HOLDS THAT ROUTE IN A LIST FORM
'''
while grid[i][j] != 1:
    
    # Check which paths are possible

    if (i+1<len(grid) and j+1<len(grid)) and (grid[i][j+1] == 1 or grid[i+1][j] == 1):
        # Reached the end of lattice path
        current_path.append((i+1, j+1))
        paths.append(current_path)
        break

    if i+1 < len(grid) and j+1 < len(grid):
        # Choose random path
        choice = np.random.randint(low=0, high=2)
        if choice == 0:
            # Move right
            current_path.append((i, j+1))
            i, j = i, j+1

        elif choice == 1:
            # Move down
            current_path.append((i+1, j))
            i, j = i+1, j

    elif i+1 >= len(grid):
        # only right path possible
        current_path.append((i, j+1))
        i, j = i, j+1

    elif j+1 >= len(grid):
        # only down path possible
        current_path.append((i+1, j))
        i, j = i+1, j
        
paths.append(current_path)
```

```python
def traverse_grid(n):
    '''
    Function that returns a random traversal path in a grid of size (n+1, n+1)
    The start vertex which is grid[0][0] is labelled as -1
    The end vertex which is grid[n][n] is labelled as 1
    '''
    
    grid = np.ndarray.tolist(np.zeros((n+1, n+1))) 
    # Set Grid start and end point as 1 for identifier
    grid[0][0] = -1
    grid[n][n] = 1
    
    i, j = 0, 0 # (i, j) represent the current matrix element coordinates
    current_path = [(i, j)]

    '''
    THE BELOW CODE FINDS A ROUTE FROM THE START TO THE END
    current_path HOLDS THAT ROUTE IN A LIST FORM
    '''
    while grid[i][j] != 1:

        # Check which paths are possible

        if (i+1<len(grid) and j+1<len(grid)) and (grid[i][j+1] == 1 or grid[i+1][j] == 1):
            # Reached the end of lattice path
            current_path.append((i+1, j+1))
            paths.append(current_path)
            break

        if i+1 < len(grid) and j+1 < len(grid):
            # Choose random path
            choice = np.random.randint(low=0, high=2)
            if choice == 0:
                # Move right
                current_path.append((i, j+1))
                i, j = i, j+1

            elif choice == 1:
                # Move down
                current_path.append((i+1, j))
                i, j = i+1, j

        elif i+1 >= len(grid):
            # only right path possible
            current_path.append((i, j+1))
            i, j = i, j+1

        elif j+1 >= len(grid):
            # only down path possible
            current_path.append((i+1, j))
            i, j = i+1, j

    return current_path
```

```python
'''
TO FIND ALL THE PATHS
ALGORITHM :
1. Take a MAX_EPOCHS number say 50
2. For each epoch iteration we take a path
3. If this path route has not been traversed before we append it to the ALL_PATHS
4. else continue traversing
'''
ALL_PATHS = []
hits = 0

while True:
    
    print(f'Paths Found : {len(ALL_PATHS)}')
    if hits > 10:
        break
        
    traversal = traverse_grid(n=20)
    if traversal not in ALL_PATHS:
        ALL_PATHS.append(traversal)
    elif traversal in ALL_PATHS:
        hits += 1
        print(f'HIT NUMBER : {hits}')
    
print(len(ALL_PATHS))
```
